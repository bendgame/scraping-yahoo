```python
i=0
sectors = {}
cant_find=[]
for sym in syms:
    print(base_url + sym +  extended_url)
    try:
        url = base_url + sym +  extended_url
        i+=1
        browser.visit(url)
        #ime.sleep(.1)
        sector_html = browser.html
        soup = BeautifulSoup(sector_html, 'html.parser')
        sector = soup.find('span', class_='Fw(600)').text
        if sector == 'Map: The United States of America, ranked from best to worst':
            cant_find.append(sym)
            print(sym + ' not fount')
        else:
            sectors.update({sym:sector})
            print(sym + ' , ' + sector)
    except(AttributeError):
        print(sym +' not found')
        cant_find.append(sym)
```

```python
df = pd.DataFrame(sectors, index = [0])
df_s = pd.DataFrame(df.stack()).reset_index()
df = df_s
df = df.rename(columns={0:'sector', 'level_1':'symbol'})
df = df[['symbol', 'sector']]
#df.to_csv(save_path)
```
